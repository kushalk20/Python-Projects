```python
import numpy as np
import pandas as pd
from pydataset import data
```

```python
cars_data = data('mtcars')
```

```python
cars_data
```

```python
#Finding the ouliers
lower = cars_data.mean() - 2*cars_data.std()
upper = cars_data.mean() + 2*cars_data.std()

#Creating a function to color the Oulier
def highlight_outlier(x):
    b = 'background-color: blue'
    r = 'background-color: red'
    g = 'background-color: green'
    df = pd.DataFrame('', index=x.index, columns=x.columns)
    for ind, col in enumerate(cars_data.columns):
        df.loc[(x[col]) > upper[ind], col] = g      # Green Color if a value is above the Confidence Interval
        df.loc[(x[col]) < lower[ind], col] = r      # Red Color if a value is below the Confidence Interval
        df.loc[(x[col]).between(lower[ind], upper[ind], inclusive='both'), col] = b   # Blue color if a value is between the CI
    return df

cars_data.style.apply(highlight_outlier, axis=None)
```
